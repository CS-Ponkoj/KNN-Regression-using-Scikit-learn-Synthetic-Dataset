```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples = 1000, n_features=3,n_informative = 3, random_state = 0)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y, random_state=0)
```

```python
KnnReg = KNeighborsRegressor(n_neighbors = 11)
```

```python
KnnReg.fit(X_train,y_train)
```

```python
print(KnnReg.predict(X_test))
```

```python
print('Test Score: {:.3f}'.format(KnnReg.score(X_test, y_test)))
```

```python
print('Train Score: {:.3f}'.format(KnnReg.score(X_train, y_train)))
```
